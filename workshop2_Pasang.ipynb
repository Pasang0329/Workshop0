```python
titanic_df=pd.read_csv("/content/drive/MyDrive/Ai/Titanic-Dataset.csv")
titanic_df.head()
```

```python
subset_df = titanic_df[['Name','Pclass','Sex','Age','Fare']].query('Pclass == 1')
subset_df.head()
```

```python
print("Mean Fare",subset_df['Fare'].mean())
print("Median Fare",subset_df['Fare'].median())
print("Max Fare",subset_df['Fare'].max())
print("Min Fare",subset_df['Fare'].min())
```

```python
print("Null values in Age:",subset_df['Age'].isnull().sum())
subset_df = subset_df.dropna(subset=['Age'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

gender_survival = titanic_df.groupby('Sex')['Survived'].mean()
print(gender_survival)

sns.barplot(x=gender_survival.index, y=gender_survival.values)
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.show
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = pd.read_csv("/content/drive/MyDrive/Ai/Titanic-Dataset.csv")
sns.catplot(data=titanic_df, x='Sex', y='Age', hue='Embarked', kind='bar')

plt.title('Survival by Gender and Embarkation Port')
plt.ylabel('Survival Rate')
plt.show()
```

```python
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'],5,labels= False )
group_survival = titanic_df.groupby(['Pclass','AgeGroup'])['Survived'].mean()
print(group_survival)
```

```python
group_survival.unstack().plot(kind='bar', stacked=True)
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Pclass and Age Group')
plt.show()
```
